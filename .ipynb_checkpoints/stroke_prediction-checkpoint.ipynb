# Stroke Prediction

### Attribute Information

This dataset is used to predict whether a patient is likely to get a stroke based
on the input parameters like gender, age, various diseases, and smoking status.
Each row in the data provides relevant information about the patient.

• **id**: unique identifier<br/>
• **gender**: “Male”, “Female” or “Other”<br/>
• **age**: age of the patient<br/>
• **hypertension**: 0 if the patient doesn’t have hypertension
1 if the patient has hypertension
1<br/>
• **heart_disease**: 0 if the patient doesn’t have any heart diseases,
1 if the patient has a heart disease<br/>
• **ever_married**: “No” or “Yes”<br/>
• **work_type**: “children”, “Govt jov”, “Never worked”,
“Private” or “Self-employed”<br/>
• **Residence_type**: “Rural” or “Urban”<br/>
• **avg_glucose_level**: average glucose level in blood<br/>
• **bmi**: body mass index<br/>
• **smoking_status**: “formerly smoked”, “never smoked”,
“smokes” or “Unknown”<br/>
• **smoking_status**: 1 if the patient had a stroke or 0 if not

###  2. Missing Value Imputation
KNN: Using Scikit-Learn to impute missing values

```python
from sklearn.impute import KNNImputer
from sklearn.model_selection import cross_val_score
```

```python
# Create KNN imputer with n_neighbors=5
imputer = KNNImputer(n_neighbors=5)

# Impute missing values in the BMI column
df = df.copy()
df['bmi'] = imputer.fit_transform(df[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi']])[:, 4]
```

### 4. Encoding Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm

data1 = df.copy()

le = LabelEncoder()
text_cols = list(data1.select_dtypes(include=['object']).columns)


l3 = []; l4 = [];
print('Label Encoder Transformation')
for i in tqdm(text_cols):
    data1[i] = le.fit_transform(data1[i])
    l3.append(list(data1[i].unique())); l4.append(list(le.inverse_transform(data1[i].unique())))
    print(i,' : ',data1[i].unique(),' = ',le.inverse_transform(data1[i].unique()))
```

Encoding categorical variables into numerical

```python
df=data1.copy()
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Separate the features (X) and the target variable (y)
X = df.drop(['id', 'stroke'], axis=1)
y = df['stroke']

# Split the data into train and test sets for each set of features
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

### 5. Modeling

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def modeling(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    metrics(y_test,pred)
    
def metrics(y_test,pred):
    accuracy = accuracy_score(y_test,pred)
    precision = precision_score(y_test,pred)
    recall = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    roc_score = roc_auc_score(y_test,pred,average='macro')
    
    print('accuracy : {0:.2f}, precision : {1:.2f}, recall : {2:.2f}'.format(accuracy,precision,recall))
    print('f1-score : {0:.2f}, auc : {1:.2f}'.format(f1,roc_score,recall))
```

### Before applying SMOTE

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
modeling(rf,X_train,X_test,y_train,y_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', random_state=42)
modeling(svm,X_train,X_test,y_train,y_test)
```

In both models before applying SMOTE, the accuracy is high (95% for both models), but the precision, recall, F1-score, and AUC are quite low. This suggests that although the models are predicting the majority class accurately, they are performing poorly in predicting the minority class. The low recall and precision indicate that the models are not effectively identifying the positive class (minority class) instances.

### 6. Feature Engineering
### After applying SMOTE to address class imbalance in the dataset.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)

# Apply SMOTE to oversample the training data
X_train_over, y_train_over = smote.fit_resample(X_train, y_train)

# Print shapes of datasets before and after SMOTE
print('Dataset shapes before applying SMOTE: ', X_train.shape, y_train.shape)
print('Dataset shapes after applying SMOTE: ', X_train_over.shape, y_train_over.shape)

# Print distribution of labels after applying SMOTE
print('Label distribution after applying SMOTE: \n', pd.Series(y_train_over).value_counts())
```

### 7. Evaluation

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
modeling(rf, X_train_over, X_test, y_train_over, y_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', random_state=42)
modeling(svm, X_train_over, X_test, y_train_over, y_test)
```

After applying SMOTE, we see a decrease in accuracy for both models, but there are significant improvements in precision, recall, F1-score, and AUC, especially for the SVM model. The increase in recall and precision suggests that the models are now better at identifying the positive class (minority class) instances. The F1-score, which considers both precision and recall, also improves, indicating a better balance between precision and recall.
